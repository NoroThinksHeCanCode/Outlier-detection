```python
import pandas as pd 
import numpy as np
from sklearn.ensemble import IsolationForest
from sklearn.datasets import load_iris
```

```python
print(sklearn.__version__)
```

```python
iris = load_iris()
X = iris.data[:, :2]
feature_names = iris.feature_names[:2]
model = IsolationForest(contamination= .05, random_state=42)
predictions = model.fit_predict(X)
```

```python
print(iris.data[:, :2])
```

```python
inliers = X[predictions == 1]
outliers = X[predictions == -1]
plt.figure(figsize=(10,6))
plt.scatter(inliers[:, 0], inliers[:, 1], c='blue', label='Inliers')
plt.scatter(outliers[:, 0], outliers[:, 1], c='red', label='Outliers')
plt.xlabel(feature_names[0])
plt.ylabel(feature_names[1])
plt.show()
```
